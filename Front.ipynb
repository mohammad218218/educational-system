### Define Student Class

```python
class student:
    def __init__(self,id,name,family):
        self.id=id
        self.name=name
        self.family=family

    def show(self):
        return(self.id,self.name,self.family)

    def insert(self):
        cursor.execute('insert into student Values(?,?,?)',self.show())
        connection.commit()


    def takecrs(self,c):
        tk = take(self.id , c.id , 0)
        tk.insert()

    def addmark(self,g):
        pass

        
    @staticmethod
    def select():
        cursor.execute('select * from student')
        L=cursor.fetchall()
        return L
    
    def average(s_family):
        cursor.execute("select grade from take where stu_id in (select id from student where family = :family) ",{'family' : s_family})
        ave=cursor.fetchall()
        s = [0]
        for i in range(len(ave)):
            a = ave[i][0]
            if a!= 0 :
                s.append(a)
            else:
                pass
        mean = sum(s) / len(s)
        return mean   
```
